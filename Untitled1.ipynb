```python
n = 41

x = numpy.linspace(0,1,n)
y = numpy.linspace(0,1,n)

D = 1/(n-1)

p = numpy.zeros((n,n))
pn = numpy.zeros((n,n))
q = numpy.zeros((n,n))
r = numpy.zeros((n,n))
Ad = numpy.zeros((n,n))

l2_target = 1e-6
```

```python
def L2_error(p, pn):
    return numpy.sqrt(numpy.sum((p - pn)**2)/numpy.sum(pn**2))
```

```python
def L1norm(new, old):
    norm = numpy.sum(numpy.abs(new-old))
    return norm
```

```python
l2norm = 1
l3norm = 1
iterations = 0
p = numpy.zeros((n,n))
q = numpy.zeros((n,n))
r = numpy.zeros((n,n))
Ad = numpy.zeros((n,n))

l2_conv = [] #aucune putain d'ide de ce  quoi a sert

#test k=0
p[0,2:]= -(1/(2*(D**2)))*(8*q[0,1:-1] - q[0,:-2]) - (3/D)
pn=p.copy()

p[1:-1,1:-1] = 0.25*(pn[1:-1,2:] + pn[1:-1, :-2] + pn[2:, 1:-1] + pn[:-2, 1:-1])

    #condition au bords pour p

p[:,0] = 0
p[:,n-1] = 0
p[n-1,:] = 0

print(p)

r[1:-1,1:-1] = p[1:-1,1:-1]*(-1)*(D**2) + 4*q[1:-1,1:-1] - q[1:-1,2:] - q[1:-1,:-2] - q[2:, 1:-1] - q[:-2, 1:-1]
d = r.copy()
rho = numpy.sum(r*r)
Ad[1:-1,1:-1] = -4*d[1:-1,1:-1]+d[1:-1,2:]+d[1:-1,:-2]+ d[2:, 1:-1] + d[:-2, 1:-1]
sigma = numpy.sum(d*Ad)

while l2norm > l2_target:
    pn = p.copy()
    p[1:-1,1:-1] = 0.25*(pn[1:-1,2:] + pn[1:-1, :-2] + pn[2:, 1:-1] + pn[:-2, 1:-1])
    #condition au bords pour p
    p[0,2:]= -(1/(2*(D**2)))*(8*q[0,1:-1] - q[0,:-2]) - (3/D)
    p[:,0] = 0
    p[:,n-1] = 0
    p[n-1,:] = 0
        
    qk = q.copy()
    rk = r.copy()
    dk = d.copy()
    
    alpha = rho/sigma

    q = qk + alpha*dk
    r = rk- alpha*Ad
        
    rhop1 = numpy.sum(r*r)
    beta = rhop1 / rho
    rho = rhop1
        
    d = r + beta*dk
    Ad[1:-1,1:-1] = -4*d[1:-1,1:-1] + d[1:-1,2:] + d[1:-1,:-2] + d[2:, 1:-1] + d[:-2, 1:-1]
    sigma = numpy.sum(d*Ad)
    
    q[:,0] = 0
    q[:,n-1] = 0
    q[n-1,:] = 0
    
    
    iterations += 1
    l2norm = L2_error(p,pn)
    l2_conv.append(l2norm)
    
    print('Number of CG iterations: {0:d}'.format(iterations))
    
print(p)
print(q)
```
