```python
n = 41

x = numpy.linspace(0,1,n)
y = numpy.linspace(0,1,n)

D = 1/(n-1)

p = numpy.zeros((n,n))
pn = numpy.zeros((n,n))
q = numpy.zeros((n,n))
r = numpy.zeros((n,n))
Ad = numpy.zeros((n,n))

l2_target = 1e-6
```

```python
def L2_error(p, pn):
    return numpy.sqrt(numpy.sum((p - pn)**2)/numpy.sum(pn**2))
```

```python
def L1norm(new, old):
    norm = numpy.sum(numpy.abs(new-old))
    return norm
```

```python
def Jacobi(pk1,pn):
    pn = p.copy()
    p[1:-1,1:-1] = .25 * (pn[1:-1,2:] + pn[1:-1, :-2] + pn[2:, 1:-1] + pn[:-2, 1:-1])
    return p,pn
```

```python
def initomega(p,pn):
    p[0,2:]= -(1/(2*(D**2)))*(8*q[0,1:-1] - q[0,:-2]) - (3/D)
    Jacobi(p,pn)

    #condition au bords pour p

    p[:,0] = 0
    p[:,n-1] = 0
    p[n-1,:] = 0
    return p,pn
```

```python
def mcg1(q,r,d,rho,sigma,Ad) :
    qk = q.copy()
    rk = r.copy()
    dk = d.copy()
    
    alpha = rho/sigma
    

    q = qk + alpha*dk
    r = rk- alpha*Ad
    
    print(q)
        
    rhop1 = numpy.sum(r*r)
    beta = rhop1 / rho
    rho = rhop1
        
    d = r + beta*dk
    Ad[1:-1,1:-1] = -4*d[1:-1,1:-1] + d[1:-1,2:] + d[1:-1,:-2] + d[2:, 1:-1] + d[:-2, 1:-1]
    sigma = numpy.sum(d*Ad)
    
    q[:,0] = 0
    q[:,n-1] = 0
    q[n-1,:] = 0
    
    
    
    return q,r,d,rho,sigma,Ad,qk
```

```python
def initpsi(r,d,Ad,rho,sigma):
    r[1:-1,1:-1] = p[1:-1,1:-1]*(-1)*(D**2) + 4*q[1:-1,1:-1] - q[1:-1,2:] - q[1:-1,:-2] - q[2:, 1:-1] - q[:-2, 1:-1]
    d = r.copy()
    rho = numpy.sum(r*r)
    Ad[1:-1,1:-1] = -4*d[1:-1,1:-1]+d[1:-1,2:]+d[1:-1,:-2]+ d[2:, 1:-1] + d[:-2, 1:-1]
    sigma = numpy.sum(d*Ad)
    
    
    return r,d,rho,sigma,Ad
```

```python
l2norm = 1
iterations = 0
a = 1.
b = 1.
p = numpy.zeros((n,n))
q = numpy.zeros((n,n))
r = numpy.zeros((n,n))
Ad = numpy.zeros((n,n))

l2_conv = [] #aucune putain d'idée de ce à quoi ça sert

#initialisation pour oméga(p) 
initomega(p,pn)


#intialisation pour psi(q)
initpsi(r,d,Ad,rho,sigma)

mcg1(q,r,d,rho,sigma,Ad)

print(q)
print("CACAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAAA")

while (a > l2_target and b > l2_target) :
    Jacobi(p,pn)
    #condition au bords pour p
    p[0,2:]= -(1/(2*(D**2)))*(8*q[0,1:-1] - q[0,:-2]) - (3/D)
    p[:,0] = 0
    p[:,n-1] = 0
    p[n-1,:] = 0
    
    #mcg pour psi
        
    mcg1(q,r,d,rho,sigma,Ad)
    
    
    iterations += 1
    l2norm = L2_error(p,pn)
    l2_conv.append(l2norm)
    
    a = L1norm(p,pn)
    b = L1norm(q,qk)
    
    print(a,b)
    
    print('Number of CG iterations: {0:d}'.format(iterations))
    
```
